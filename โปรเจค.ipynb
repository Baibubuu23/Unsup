ชุดข้อมูล ***Coffee Quality Data (CQI May-2023)*** นี้มีคอลัมน์เกี่ยวกับกาแฟจากฟาร์มและภูมิภาคต่าง ๆ ดังนี้

ID: รหัสประจำแต่ละรายการ

Country of Origin : ประเทศที่ปลูกกาแฟ

Farm Name : ชื่อของฟาร์มที่ปลูกกาแฟ

Lot Number : รหัสเฉพาะสำหรับล็อตกาแฟ

Mill : สถานที่หรือบริษัทที่แปรรูปกาแฟ

ICO Number : เลขจากองค์การกาแฟสากล

Company : บริษัทที่รับผิดชอบกาแฟ

Altitude : ความสูงจากระดับน้ำทะเลของพื้นที่ปลูกกาแฟ

Region : ภูมิภาคที่ปลูกกาแฟภายในประเทศ

Producer : ผู้ผลิตหรือองค์กรที่ปลูกกาแฟ

Number of Bags : จำนวนกระสอบกาแฟทั้งหมด

Bag Weight : น้ำหนักของแต่ละกระสอบกาแฟ

In-Country Partner : พันธมิตรท้องถิ่นที่มีส่วนร่วมในการผลิตกาแฟ

Harvest Year : ปีที่เก็บเกี่ยวกาแฟ

Grading Date : วันที่ประเมินคุณภาพกาแฟ

Variety : สายพันธุ์ของเมล็ดกาแฟ

Status : สถานะของกาแฟ เช่น แปรรูป ส่งออก

Processing Method : วิธีที่ใช้ในการแปรรูปเมล็ดกาแฟ เช่น การล้าง การตากแห้ง

 ***Aroma, Flavor, Aftertaste, Acidity, Body, Balance : กลิ่น, รสชาติ, รสหลัง, ความเป็นกรด, เนื้อกาแฟ, ความสมดุล***

Uniformity, Clean Cup, Sweetness, Overall : ความสม่ำเสมอ, ความสะอาดของถ้วย, ความหวาน, คะแนนรวม

Defects : จำนวนข้อบกพร่องที่พบในเมล็ดกาแฟ

Total Cup Points : คะแนนรวมที่ได้รับจากการประเมินคุณภาพกาแฟ

Moisture Percentage : เปอร์เซ็นต์ความชื้นในเมล็ดกาแฟ

Category One Defects : ข้อบกพร่องที่ส่งผลต่อคุณภาพของเมล็ดกาแฟ

Quakers : เมล็ดกาแฟที่ไม่สามารถคั่วได้ดี

Color : สีของเมล็ดกาแฟ

Category Two Defects : ข้อบกพร่องเล็กน้อยในเมล็ดกาแฟ

Expiration : วันที่หมดอายุของกาแฟ

Certification Body : องค์กรที่รับรองคุณภาพของกาแฟ

Certification Address : ที่อยู่ของหน่วยงานรับรอง

Certification Contact : ข้อมูลการติดต่อของหน่วยงานรับรอง
