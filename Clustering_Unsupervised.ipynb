```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
#(['data', 'target', 'target_names', 'DESCR', 'feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_data =scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
PCA(n_components=2)
```

```python
x_pca = pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
(569, 30)
(569, 2)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='prism')
plt.xlabel('First Principle Component')
plt.ylabel('Second Principle Component')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming x_pca is already calculated using PCA
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], cmap='prism')  # Removed the target variable
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA (Unsupervised)')
plt.show()
```

```python
from sklearn.cluster import KMeans

# K-Means Clustering
kmeans = KMeans(n_clusters=2)  # สมมติว่าเราต้องการ 2 กลุ่ม
kmeans.fit(x_pca)
clusters = kmeans.labels_

# Visualization
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=clusters, cmap='prism')  # Color based on K-Means clusters
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA with K-Means Clustering (Unsupervised)')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# สมมติว่าคุณมี DataFrame ที่ชื่อว่า 'df' ที่มีข้อมูล 32 คอลัมน์
# เราต้องทำการ Standardize ข้อมูลก่อน เพราะ PCA และ K-Means อ่อนไหวต่อสเกลของข้อมูล
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# 1. ทำ PCA เพื่อการลดมิติ
pca = PCA(n_components=2)  # เลือกจำนวนคอมโพเนนต์ที่ต้องการ (เช่น 2 สำหรับการ visualization)
df_pca = pca.fit_transform(df_scaled)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(df_pca)

labels = kmeans.labels_

plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title('K-Means Clustering on PCA-Reduced Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# สร้างโมเดล DBSCAN
dbscan = DBSCAN(eps=0.7, min_samples=10)  # ลองลดค่า eps และปรับค่า min_samples
labels_dbscan = dbscan.fit_predict(df_pca)

# แสดงผลการจัดกลุ่มใหม่
plt.figure(figsize=(10, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels_dbscan, cmap='plasma', marker='o', edgecolor='k', s=50)
plt.title('DBSCAN Clustering on PCA-reduced Data (Adjusted eps and min_samples)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# ใช้ข้อมูลที่ผ่านการทำ PCA แล้ว (df_pca)
# df_pca ควรเป็น array 2D ที่มีข้อมูลลดมิติลงแล้ว เช่น (n_samples, 2)

# สร้าง linkage matrix สำหรับ hierarchical clustering
Z = linkage(df_pca, method='ward')  # ใช้วิธี 'ward' สามารถเปลี่ยนเป็น 'single', 'complete', 'average' ได้

# สร้าง dendrogram เพื่อแสดงการจัดกลุ่มแบบ hierarchical
plt.figure(figsize=(10, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.axhline(y=60, color='r', linestyle='--')  # เส้นตัดเพื่อควบคุมการแบ่งกลุ่ม
plt.show()

# ใช้ Agglomerative Clustering เพื่อตัด dendrogram และสร้างกลุ่ม
num_clusters = 2 # กำหนดจำนวนกลุ่มที่ต้องการ
hierarchical_clustering = AgglomerativeClustering(n_clusters=num_clusters, linkage='ward')
labels_hierarchical = hierarchical_clustering.fit_predict(df_pca)

# แสดงผลการจัดกลุ่ม
plt.figure(figsize=(10, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels_hierarchical, cmap='plasma', marker='o', edgecolor='k', s=50)
plt.title(f'Hierarchical Clustering on PCA-reduced Data (n_clusters={num_clusters})')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.decomposition import PCA
import pandas as pd

# สมมติว่าคุณมี DataFrame ที่ชื่อ df ซึ่งเป็นข้อมูลจริง
# df = pd.read_csv('your_dataset.csv')  # โหลดข้อมูลของคุณที่นี่



# ขั้นตอนที่ 2: ทำการจัดกลุ่มแบบ divisive (ใช้ linkage)
Z = linkage(df_pca, method='ward')

# ขั้นตอนที่ 3: สร้าง dendrogram
plt.figure(figsize=(12, 8))
dendrogram(Z)
plt.title('Dendrogram for Divisive Clustering')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.axhline(y=8000, color='r', linestyle='--')  # เส้นตัดเพื่อควบคุมการแบ่งกลุ่ม
plt.show()

# ขั้นตอนที่ 4: กำหนดจำนวนกลุ่มที่ต้องการ (2 กลุ่ม)
num_clusters = 2
labels_divisive = fcluster(Z, num_clusters, criterion='maxclust')

# ขั้นตอนที่ 5: แสดงผลการจัดกลุ่ม
plt.figure(figsize=(10, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels_divisive, cmap='plasma', marker='o', edgecolor='k', s=50)
plt.title(f'Divisive Clustering on Real Data (n_clusters={num_clusters})')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import pandas as pd


# ขั้นตอนที่ 3: สร้างโมเดล GMM และปรับพารามิเตอร์
num_clusters = 2  # กำหนดจำนวนกลุ่มที่ต้องการ
gmm = GaussianMixture(n_components=num_clusters)
gmm.fit(df_pca)

# ขั้นตอนที่ 4: ทำนายกลุ่ม
labels_gmm = gmm.predict(df_pca)

# ขั้นตอนที่ 5: แสดงผลการจัดกลุ่ม
plt.figure(figsize=(10, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels_gmm, cmap='plasma', marker='o', edgecolor='k', s=50)
plt.title(f'GMM Clustering on Real Data (n_clusters={num_clusters})')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
