```python
from sklearn import datasets

iris = datasets.load_iris()
x = iris.data
y = iris.target
x_index = 0  # Sepal Length
y_index = 1  # Sepal Width
x = iris.data[:, x_index]
y = iris.data[:, y_index]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=iris.target, cmap='viridis', s=50, alpha=0.8)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.title('Scatter plot of Iris Dataset')
plt.colorbar(label='Target', ticks=[0, 1, 2])
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(iris.data)
x = principal_components[:, 0]
y = principal_components[:, 1]
```

```python
from matplotlib.colors import ListedColormap
single_color_map = ListedColormap(['black'])
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=iris.target, cmap=single_color_map, s=50, alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter plot of Iris Dataset')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=50, alpha=0.8)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Scatter plot of PCA Components (PCA1 vs PCA2)')
plt.show()
```

```python
import numpy as np
pca = PCA()
pca.fit(iris.data)
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(8, 6))
plt.plot(np.arange(1, 5), explained_variance_ratio, marker='o', linestyle='--')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot of PCA Components')
plt.xticks(np.arange(1, 5), ['PC1', 'PC2', 'PC3', 'PC4'])
plt.show()
```
